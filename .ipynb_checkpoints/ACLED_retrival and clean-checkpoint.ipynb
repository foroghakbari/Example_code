```python
# Set API endpoint and parameters
endpoint= "https://api.acleddata.com/acled/read/?key=GsdDQ4dqB12pDDOdS1Z7&email=forogh.akbari.4726@student.uu.se"
params = {
    'event_type': 'Battles',
    'start_date': '2014-01-01',
    'end_date': '2023-01-01',
    'limit': 1000, # Default limit is 500 rows
    'page': 1 # Start with the first page
}

# Send request and retrieve data
data = []
while True:
    response = requests.get(endpoint, params=params)
    page_data = response.json()['data']
    if not page_data:
        break
    data.extend(page_data)
    params['page'] += 1
    # Use limit query name to change number of rows per page
    # For example, params['limit'] = 1000 for export_type='monadic'
    # Setting limit as 0 will return all relevant data, but may cause a timeout error

# Convert data to DataFrame
df = pd.DataFrame(data)
```
